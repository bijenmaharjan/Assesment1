```python
# 1
q1=Sa_df['Composite score'].quantile(0.25)
q3=Sa_df['Composite score'].quantile(0.75)
iqr=q3-q1
lower_value=q1-1.5*iqr
upper_value=q3+1.5*iqr
outliers_cs = Sa_df[(Sa_df['Composite score'] < lower_value) | (Sa_df['Composite score'] > upper_value)][['Country name','Composite score']]
print(outliers_cs)
```

```python
# 1
q1=Sa_df['Log GDP per capita'].quantile(0.25)
q3=Sa_df['Log GDP per capita'].quantile(0.75)
iqr=q3-q1
lower_value=q1-1.5*iqr
upper_value=q3+1.5*iqr
outliers_gdp = Sa_df[(Sa_df['Log GDP per capita'] < lower_value) | (Sa_df['Log GDP per capita'] > upper_value)][['Country name','Log GDP per capita']]
print(outliers_gdp)
```

Define outliers using the 1.5 Ã— IQR rule.
1.5 is exactly used in outliers because it is more appropriate value to count  outlier. 1.5 ignores all the small difference so outliers can actually be calculated.

```python
# 3
Q1_score = Sa_df['score'].quantile(0.25)
Q3_score = Sa_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = Sa_df['Log GDP per capita'].quantile(0.25)
Q3_gdp = Sa_df['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp
# 2. Define outliers using the 1.5 Ã— IQR rule
outliers_s = Sa_df[(Sa_df['score'] < (Q1_score - 1.5 * IQR_score)) |
                                (Sa_df['score'] > (Q3_score + 1.5 * IQR_score))]

gdp_outliers = Sa_df[(Sa_df['Log GDP per capita'] < (Q1_gdp - 1.5 * IQR_gdp)) |
                               (Sa_df['Log GDP per capita'] > (Q3_gdp + 1.5 * IQR_gdp))]

outliers = pd.concat([outliers_s, gdp_outliers]).drop_duplicates()

# 3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=Sa_df, color='blue', label='Composite Score')
sns.scatterplot(x='Log GDP per capita', y='score', data=outliers_s, color='brown', label='original score')
plt.title('Scatter Plot: GDP per Capita vs Score with Outliers Highlighted')
plt.xlabel('Log GDP per capita')
plt.ylabel('Score')
plt.legend()

# Show the plot
plt.show()

```
